**Побудуємо графік, який показує відображення від довжини слова до частоти слова:**

```python
word_length_freq = {}
for word, freq in fdist.items():
    word_length = len(word)
    if word_length not in word_length_freq:
        word_length_freq[word_length] = freq
    else:
        word_length_freq[word_length] += freq
```

```python
plt.figure(figsize=(12, 8))
plt.bar(word_length_freq.keys(), word_length_freq.values(), color='black')
plt.title('Відношення довжини до частоти')
plt.xlabel('Довжина слова')
plt.ylabel('Частота')
plt.show()
```

```python
import string
from collections import Counter
```

```python
my_text_str = ' '.join(my_text)
```

```python
sentences = sent_tokenize(my_text_str)
```

```python
translator = str.maketrans('', '', string.punctuation)
```

```python
last_words = []
for sentence in sentences:
    words = word_tokenize(sentence.translate(translator))
    if len(words) > 0:
        last_word = words[-1].lower()
        last_words.append(last_word)
```

```python
word_counts = Counter(last_words)
```

```python
top_10_words = word_counts.most_common(10)
print(top_10_words)
```

Аналіз тексту Ліни Костенко "Маруся Чурай"

```python
with open('kostenko-lina-vasylivna-marusia-churay.txt', 'r', encoding='windows-1251') as file:
    raw_text = file.read()

raw_text = raw_text.translate(str.maketrans('', '', string.punctuation)).lower()
tokens = nltk.word_tokenize(raw_text)
text = nltk.Text(tokens)
freq_dist = FreqDist(text)
```

```python
text
```

Все, тепер текст підготовлений, аналізуємо

```python
filtered_words = [word for word in text if len(word) > 3]
freq_dist_filtered = FreqDist(filtered_words)
top_20_words = freq_dist_filtered.most_common(20)

print("Топ-20 слів з тексту, які мають довжину більше 3 літер:")
for word, frequency in top_20_words:
    print(f"{word}: {frequency}")
```

```python
trigrams = list(ngrams(tokens, 2))
freq_dist_trigrams = FreqDist(trigrams)
top_20_trigrams = freq_dist_trigrams.most_common(20)

print("Топ-20 триграм з тексту:")
for trigram, frequency in top_20_trigrams:
    print(f"{' '.join(trigram)}: {frequency}")
```

```python
trigrams_full = [trigram for trigram in ngrams(tokens, 3) if any(len(word) > 3 for word in trigram)]
freq_dist_trigrams_full = FreqDist(trigrams_full)
top_20_trigrams_full = freq_dist_trigrams_full.most_common(20)


print("Топ-20 триграм з тексту, де принаймні одне слово має довжину більше 3:")
for trigram, frequency in top_20_trigrams_full:
    print(f"{' '.join(trigram)}: {frequency}")
```

```python
def extract_money(text):
    money_amounts = re.findall(r'\b\d+(?:\.\d+)?\b', text)
    money_amounts = [float(amount) for amount in money_amounts]
    return money_amounts
```

```python
money_amounts = extract_money(text)
print("Отримані грошові суми:", money_amounts)
print("Загальна сума:", sum(money_amounts))
```

```python
import json

def process_json(json_data):
    json_data['Config']['Nodes'] = [node for node in json_data['Config']['Nodes'] if not node.startswith('8.')]

    server_pattern = r'(.+)\.(.+)'
    json_data['Config']['Server'] = re.sub(server_pattern, r'{"name": "\1", "type": "\2"}', json_data['Config']['Server'])

    return json_data
```

```python
json_string = '''
{
    "HTTPHost": "localhost",
    "HTTPPort": 8545,
    "Config": {
        "Enabled": true,
        "Server": "wakuv2.prod",
        "Nodes": ["8.210.222.231", "168.166.125.145"]
    }
}
'''
```

```python
json_data = json.loads(json_string)
processed_json = process_json(json_data)
print(json.dumps(processed_json, indent=4))
```
